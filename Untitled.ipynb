```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("matches.csv")
```

```python
df=df[(df['team1']=='Mumbai Indians')|(df['team1']=='Delhi Daredevils')|(df['team1']=='Delhi Capitals')]
df=df[(df['team2']=='Mumbai Indians')|(df['team2']=='Delhi Daredevils')|(df['team2']=='Delhi Capitals')]
df['winner'].value_counts()
```

```python
df['team1'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df['team2'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df['toss_winner'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df['winner'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df.head()
```

```python
df=df[(df['team1']=='Mumbai Indians')|(df['team1']=='Delhi Capitals')]
df=df[(df['team2']=='Mumbai Indians')|(df['team2']=='Delhi Capitals')]
```

```python
df['winner'].value_counts()
```

```python
tw=df['toss_winner'].value_counts()
tw
```

```python
df['toss_decision'].value_counts()
```

```python
tww=df[df['toss_winner']==df['winner']]['winner'].value_counts()
tww
```

OUT OF 14 TIMES, DELHI WON ONLY 5 TIMES WHEN THEY WON THE TOSS. 
OUT OF 10 TIMES, MUMBAI WON ONLY 3 TIMES WHEN THEY WON THE TOSS.

CHANCES OF WINNNG OF BOTH THE TEAMS AFTER WINNING TOSS IS NOT THAT GOOD. 

```python

win_percent=(tww/tw)*100
win_percent
```

```python
data=df[df['toss_winner']=='Mumbai Indians']['toss_decision'].value_counts()
data
```

```python
fig=plt.figure(figsize =(10, 7)) 
plt.pie(data,labels=['bat','field'],colors=['cyan','indigo'],startangle=90,autopct='%1.2f%%')
plt.show()
```
